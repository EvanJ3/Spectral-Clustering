## Spectral Clustering

In this project, we will implement and discuss a pure python implementation of spectral clustering for graph analysis. We will study a political blogs dataset first compiled for the paper Lada A. Adamic and Natalie Glance, The political blogosphere and the 2004 US Election, in Proceedings of the WWW-2005 Workshop on the Weblogging Ecosystem (2005). The dataset nodes.txt contains a graph with $n = 1490$ vertices (nodes) corresponding to political blogs. Each vertex has a 0-1 label (in the 3rd column) corresponding to the political orientation of that blog. We will consider this as the true label and try to reconstruct the true label from the graph using the spectral clustering on the graph. The dataset edges.txt contains edges between the vertices. In the context of k means, we use the value of k to represent a number of clusters each of which is focused on identifying similarity among observations via inter-cluster distance minimization or similarity. However, in the context of spectral clustering our choice of k although equally as important a choice as in k means, takes on a different meaning. Unlike k means, in spectral clustering we are using graphs as input which measure connectivity. To this end, spectral clustering attempts to identify densely connected groups of nodes (communities) within the graph as a whole. Thus, in the context of spectral clustering, the value of k represents the number of communities we wish to group the graph into.  


---

Using spectral clustering we will find the $k = 2$, $k = 3$, and $k = 4$ clusters in the network of political blogs (each node is a blog, and their edges are defined in the file edges.txt). We will treat the network as an un-directed graph; thus, when constructing the adjacency matrix, we must be sure to make sure it is symmetrical. Then we will report the majority labels in each cluster, when $k = 2, 3, 4$, respectively.

My implementation of the spectral clustering algorithm is contained in repository under the name "spectral\_clustering.py". The images below show the different true labels and their respective counts are contained within each cluster over k values 2,3, and 4.

The below code cell imports the custom module created to run the Spectral Clustering algorithm as well as other necessary dependencies.

```python
import time
import pandas as pd
import numpy as np
import os,sys,os.path,shutil,platform
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import custom_spectral as cs
import warnings
```

### Forming The Adjacency Matrix:

```python
edges = pd.read_table('edges.txt',names=['vertex_1','vertex_2'])
nodes = pd.read_table('nodes.txt',names=['vertex_1','Connected','vertex_2']).reset_index()
nodes = nodes['Connected']
reduced_adj_matrix, reduced_nodes_array = cs.form_reduced_adjancency_and_labels(edges,nodes)

plt.spy(reduced_adj_matrix)
plt.title('Adjancency Matrix')
plt.show()
```

### Forming The Laplacian Matrix:

```python
laplacian_matrix = cs.adj_to_laplacian(adj_matrix=reduced_adj_matrix,return_d_matrix=False)
plt.spy(laplacian_matrix)
plt.title("Lapacian")
plt.show()
```

### Displaying Graphed Real Parts of Eigenvalues by Ascending Magnitude:

```python
eig_vals,eig_vectors = scipy.linalg.eig(laplacian_matrix)
idx_e = eig_vals.argsort()[::1]
eig_vals = eig_vals[idx_e]
eig_vectors = eig_vectors[:,idx_e].real
plt.plot(eig_vals)
plt.show()
```

## Cluster Assignements for Various K Values:

Of additonal note on the implementation, if the notebook is re-run the output can change due to random initalization. The user can specify a numpy seed to make the output reproducable. ### Spectral Clustering with K = 2

```python
k_2_mis_matches = []
k=2
print('--------K=%s Case --------'%k)
k2_labels = cs.K_means_Spectral(eig_vectors,reduced_nodes_array,k=k,iter_max=200)
for i in np.unique(k2_labels):
    indicies = np.argwhere(k2_labels==i)[:,0]
    true_labs = reduced_nodes_array[indicies]
    mode,counts = scipy.stats.mode(true_labs)
    mis_match_rate = (true_labs.shape[0]-counts)/true_labs.shape[0]
    true_labs_series = pd.Series(reduced_nodes_array[indicies])
    print('--------Cluster %s --------'%i)
    #print('Cluster Mis-Match Rate %s'% round(mis_match_rate[0],2))
    print(true_labs_series.value_counts())
```

As we can see from the above output, the majority label for cluster 0 is 1, and the majority label for cluster 1 is 0.

### Further Discussion:

In looking over some of our results above it is notable that the algorithm for at least the reasonably low values of k tested performed poorly in the context of overall mis-match rate. However, we can still see some interesting properties of spectral clustering given the limited output above. Consider how in each case the algorithm creates a few distinct clusters of only a single or few points then clumping the rest of the data into one large messy or less dense cluster. This indicates based on our knowledge of spectral clustering implementation that these small clusters are very heavily interconnected far more connected than the rest of the dataset at large. This indicates these points could potentially be outliers if as we scale up k the behavior persists while other clusters remain large. Alternatively, these could just be the first easily identifiable community, and as we expand k we will see similar small communities emerge in more isolated clusters with the misfits being lumped into one or a few larger less connected clusters. If that is the case we will see certain values of k that permit clean or reasonably clear separability. 

Now let's tune $k$ and find the number of clusters to achieve a reasonably small it mismatch rate. In searching for the proper tuning the k value I decided to, for several values of k, run the algorithm with a set max number of iterations (500) constant for each k value, and supply the same initialization centers by setting a constant random seed initialization for each value of k. Then, I used the output of these runs from k values of 0 to 200 to plot each choice of k's overall mismatch rate alongside one another. As we can clearly see from the output graph below there appears to be a consensus on mis-match rate for most values of k. However, there are certain values for k which we can see strong drops or equivalently increases in accuracy in some cases to very high accuracy. This indicates to an observer that although the clusters don't always agree perfectly there are certain values of k which separate the dataset nicely into agreeable clusters quite well. This behavior persists over many differnt random initializations, but at slightly different values of k.These values of k will likely require further investigation. Additionally, it is notably that because we are dealing with binary label data, mis-match rates are somewhat excessive in that the error is double counted with a swapped label which helps explain the erratic behavior of the charted values below. This opens the door to employ different accuracy metrics to further explore the data in future analysis.
